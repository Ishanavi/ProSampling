```python
import pandas as pd
import csv
import plotly.figure_factory as ff
import random
import statistics
```

```python
with open("medium_data.csv") as f:
  reader = csv.reader(f)
  data_list = list(reader)

data_list.pop(0)
response_list = []
clap_data = []
read_data = []

for i in data_list:
 clap_data.append(float(i[5]))
 response_list.append(i[6])
 read_data.append(float(i[7]))


fig = ff.create_distplot(
                            [clap_data],
                            ["Clap"],
                            show_hist = False
)

fig.show()

fig = ff.create_distplot(
                            [read_data],
                            ["Reading Time"],
                            show_hist = False
)

fig.show()
```

```python
def randomSetOfMean(counter):
  dataset = []
  for i in range(0,counter):
        random_index = random.randint(0,len(clap_data))
        value = clap_data[random_index]
        dataset.append(value)
  mean = statistics.mean(dataset)
  return mean
  
```

```python
def showFig(mean_list):
  fig = ff.create_distplot(
                            [mean_list],
                            ["Claps Received"],
                            show_hist = False
  )
  fig.show()
```

```python
def setupClap():
  mean_list = []
  for i in range(0,100):
    set_of_mean = randomSetOfMean(30)
    mean_list.append(set_of_mean)
  showFig(mean_list)
  mean = statistics.mean(mean_list)
  print("Sampling Mean --> ",mean)
setupClap()
```
